```python
%matplotlib notebook

import geopandas as gpd
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
import geoplot


sql = "select * from gadm36_ind_1"

con = psycopg2.connect(database="ok", user="postgres", password="",host="localhost")

df = gpd.GeoDataFrame.from_postgis(sql, con, geom_col='wkb_geometry')

data = pd.read_csv('/home/cryptik/CSV')

dat = data['TOTAL'].tolist()

df['TOTAL'] = data['TOTAL']

fig, ax = plt.subplots(figsize  = (8, 8))

geoplot.choropleth(df, hue='TOTAL', cmap='PuBuGn', figsize=(8, 4),legend='True',ax=ax,scheme='quantiles')

ax.set_axis_off()

plt.title("LPG Consumers")

```
